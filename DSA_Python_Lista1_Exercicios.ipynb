#### Exercício 3

Qual é a principal diferença entre uma variável de escopo local e uma de escopo global?



*   variável de escopo **LOCAL**

1.   Criada dentro de uma função ou bloco de código {}, onde pode ser vista.
2.   Criada quando a função inicia e detruída quando termina.
3.   Erros: BAIXO


*   variável de escopo **GLOBAL**


1.   Fora de todas as funções, geralmente no topo do arquivo, visto por qualquer parte do programa
2.   Existe enquanto o programa estiver em execução
3.   Erros: ALTO

Crie uma variável chamada saldo com o valor 500.50 (float). Em seguida, crie uma variável saque com o valor 200.25 (float). Subtraia o saque do saldo e imprima o saldo final formatado para duas casas decimais.

```python
# Solução
# --- VARIÁVEL GLOBAL ---
# Definida fora de qualquer função
TAXA_DOLAR = 5.50

def converter_para_real(quantidade_dolar):
    # A função consegue ler a variável global sem problemas
    total = quantidade_dolar * TAXA_DOLAR
    return total

def exibir_cotacao():
    # Outra função acessando a mesma variável global
    print(f"A cotação atual utilizada é: R$ {TAXA_DOLAR}")

# Executando o programa
exibir_cotacao()
meu_valor = 100
print(f"US$ {meu_valor} equivalem a R$ {converter_para_real(meu_valor)}")

# --- VARIÁVEL LOCAL---
# Definida dentro da função

def calcular_preco_final(preco_base):
    # --- VARIÁVEL LOCAL ---
    # 'valor_desconto' só existe aqui dentro
    taxa = 0.15
    valor_desconto = preco_base * taxa

    preco_final = preco_base - valor_desconto
    return preco_final

# Executando a função
print(f"Preço com desconto: R$ {calcular_preco_final(100)}")

# Tentar acessar a variável local fora da função causará um ERRO:
# print(valor_desconto)  # NameError: name 'valor_desconto' is not defined
```

#### Exercício 30

Escreva um programa que peça ao usuário para digitar sua altura em metros (ex: 1.75) e seu peso em quilogramas (ex: 68.5). Calcule o Índice de Massa Corporal (IMC) usando a fórmula IMC = peso / (altura * altura) e imprima o resultado formatado com duas casas decimais.

```python
# Solução
altura_str = input("Digite sua altura (ex: 1.75):")
peso_str = input("Digite seu peso (ex: 68.5):")

# Substitui vírgula por ponto para garantir a conversão correta para float
altura = float(altura_str.replace(',', '.'))
peso = float(peso_str.replace(',', '.'))

imc = float(peso) / (float(altura)*float(altura))
print(f"Seu ìndice de Massa Corporal é: {imc}")
```
